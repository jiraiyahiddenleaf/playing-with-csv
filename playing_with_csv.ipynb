### 2. opening csv file from URL

### 12. handling dates

    sometimes the dates are converted into string (object) so in this case we can't use the date functionality 

```python
pd.read_csv('IPL Matches 2008-2020.csv').info()
# date is in object 
```

```python
pd.read_csv('IPL Matches 2008-2020.csv', parse_dates=['date']).info()
# now the date is in datetime64
```

### 13. Convertors 

```python
pd.read_csv('IPL Matches 2008-2020.csv').sample(10)
# here team name is too big so we will rename them 
```

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    if name == "Kolkata Knight Riders":
        return "KKR"
    if name == "Rajasthan Royals":
        return "RR"
    if name == "Chennai Super Kings":
       return "CSK"
    else:
        return name
    
    # we can change all the names of the team 
```

```python
rename("Chennai Super Kings")
```

```python
#rename the teams 
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename, 'team2':rename}).sample(5)
```

### 15. Loading a huge dataset in chunks

 In pandas, chunks refer to a technique used when handling large datasets, where the data is divided into smaller, more manageable pieces called chunks. This helps to reduce memory usage and allows for efficient processing and iteration through the data.

```python
dfs = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
for chunks in dfs:
    print(chunks.shape)
```
